```python
msno.matrix(df)
plt.show()
```

```python
#Imputación de datos faltantes: host_nombre. Faltantes: 6
#Ya que no hay forma de predecir el nombre o imputarlo por estadística, se
#imputa con el enunciado "No registrado"
df['host_nombre'] = df['host_nombre'].fillna('No registrado')
```

```python
#Imputación de datos faltantes: host_desde. Faltantes: 6
#Visualizar las filas con datos faltantes en 'host_desde'
display(df[df['host_desde'].isnull()])
```

```python
# Imputación de datos faltantes: host_desde con valores de primer_review
#Al ver las filas se pudo observar que primer review estaba presente y se puede
#tomar este dato para imputar host_desde
df['host_desde'] = df['host_desde'].fillna(df['primer_review'])
```

```python
#Imputación de datos faltantes: descripción. Faltantes: 1142
#Ya que esta variable da información especializada de cada propiedad, solo se
#imputa con el enunciado "Sin descripción"
df['descripción'] = df['descripción'].fillna('Sin descripción')
```

```python
#tiempo_respuesta	2799
#Checamos valores únicos de tiempo_respuesa
print(df['tiempo_respuesta'].unique())
```

```python
#Ya que es una variable categórica, creamos una nueva categoría de "Sin respuesta" y la asignamos a los valores nulos.
df['tiempo_respuesta'] = df['tiempo_respuesta'].fillna('Sin respuesta')
```

```python
#tasa_respuesta	2799
#Ya que tasa_repuesta tiene los mismos faltantes, se asigna un porcentaje de 0%
df['tiempo_respuesta'] = df['tiempo_respuesta'].fillna('0%')
```

```python
#tasa_aceptación_host	2187
#Puede considerarse ausencia de respuesta y así como otros registros, se le asgina el 0%
df['tasa_aceptación_host'] = df['tasa_aceptación_host'].fillna('0%')
```

```python
#superhost	674
#Es una variable ambigüa, en cambio, creamos una categoría para decir que no se contestó
df['superhost'] = df['superhost'].fillna('n/c')
```

```python
#host_vecindario	8085
#Esta variable puede cambiar mucho conforme al host y no hay forma de predecirla, se le asigna
#"no especificado"
df['host_vecindario'] = df['host_vecindario'].fillna('No especificado')
```

```python
#host_listados	6
#Esta es la cantidad de listados que tiene el host, ya que aparece por lo menos una propiedad en el dataset,
#Reemplazamos los faltantes con el número 1
df['host_listados'] = df['host_listados'].fillna(1)
```

```python
#host_contacto	6
#Vemos los valores únicos
print(df['host_contacto'].unique())
```

```python
#Rellenamos con "Sin contacto"
#También se puede observar un registro '[]', este lo eliminamos y lo transformarmos en NA

df['host_contacto'] = df['host_contacto'].replace('[]', np.nan)
df['host_contacto'] = df['host_contacto'].fillna('Sin contacto')
```

```python
#host_foto_perfil	6
print(df['host_foto_perfil'].unique())
```

```python
#Reemplazamos los nan con f, intuyendo que estos valores se deben a ausencia de foto
df['host_foto_perfil'] = df['host_foto_perfil'].fillna('f')
```

```python
#host_identidad_verif	6
print(df['host_identidad_verif'].unique())
```

```python
#Igualmente reemplazamos faltantes con "f"
df['host_identidad_verif'] = df['host_identidad_verif'].fillna('f')
```

```python
#baños  4545
# Calcular el promedio de baños por tipo de propiedad
avg_baños_por_propiedad = df.groupby('tipo_propiedad')['baños'].mean().sort_values(ascending=False)
```

```python
# Imputar los valores faltantes en 'baños' con el promedio por tipo de propiedad
df['baños'] = df.groupby('tipo_propiedad')['baños'].transform(lambda x: x.fillna(x.mean()))
```

```python
#Después de imputar con la media por tipo de propiedad, todavía quedaron valores faltantes, estos son por su tipo único de habitación
#Visualizamos los faltantes
df[df['baños'].isnull()]
```

```python
# Imputar los valores faltantes en 'baños' con los valores de 'recámaras'
df['baños'] = df['baños'].fillna(df['recámaras'])
```

```python
#recámaras	616
avg_baños_por_propiedad = df.groupby('tipo_propiedad')['baños'].mean()
avg_baños_por_propiedad
```

```python
#camas	4574
avg_camas_por_propiedad = df.groupby('tipo_propiedad')['camas'].mean().sort_values(ascending=False)
avg_camas_por_propiedad
```

```python
df['camas'] = df.groupby('tipo_propiedad')['camas'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Rellenar los valores NaN de 'camas' con la columna 'alojados' dividida por 2
df['camas'] = df['camas'].fillna(df['alojados'] / 2)
```

```python
#Reemplazamos por el precio promedio por tipo de propiedad
df['precio'] = df.groupby('tipo_propiedad')['precio'].transform(lambda x: x.fillna(x.mean()))
```

```python
#Transformamos a númerico, quitando signos y puntos
df['precio'] = df['precio'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Reemplazar los NaN de precio por la media de acuerdo al tipo de habitación
df['precio'] = df.groupby('tipo_habitación')['precio'].transform(lambda x: x.fillna(x.mean()))
```
